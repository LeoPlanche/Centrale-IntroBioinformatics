We plot the FST matrix, what do you notice? Give an interpretation of the results.

Note: If you do not have zeros on diagonal, there is a mistake!

```python
import matplotlib.pyplot as plt

def plot_fst(FST,cell_size=1, map = {}):
    if map == {}:
        keys = sorted(FST.keys())
    else:
        keys = sorted(map, key=map.get)


    plt.figure(figsize=(cell_size*len(keys), cell_size*len(keys)))
    # Create a 2D numpy array
    matrix = np.array([[round(FST[row][col],2) for col in keys] for row in keys])

    # Plot heatmap
    plt.imshow(matrix, cmap="viridis")
    #plt.colorbar()

    # Set ticks and labels
    plt.xticks(range(len(keys)), keys)
    plt.yticks(range(len(keys)), keys)

    # Optionally add values on cells
    for i in range(len(keys)):
        for j in range(len(keys)):
            plt.text(j, i, str(matrix[i, j]), ha='center', va='center', color='white')
    plt.tight_layout()
    plt.show()
    
# TODO uncomment the next line...
#plot_fst(FST)
```
